```python
import os
import json
import pandas as pd
```

```python
file_path = '/home/moose/Documents/python/la_ethics/covid_data'
file_ext = 'json'
list_of_ext_files = []
for root, dirs, files in os.walk(file_path):
    for file in files:
        if file.endswith(file_ext):
             list_of_ext_files.append(os.path.join(root, file))
```

```python
df = pd.DataFrame()
for file_path in list_of_ext_files:
    with open(file_path, 'r') as infile:
        jf = json.load(infile)
    arcgis_features = [jf['features'][num]['attributes'] for num in range(0, len(jf['features']))]
    df = pd.concat([df, pd.DataFrame.from_dict(arcgis_features)], ignore_index=True)
    
df = df.sort_values(by='OBJECTID', ascending=True)
```
